```python
import pandas as pd
import os
import random
from PIL import Image
```

```python
df = pd.read_excel("E:\\dataset\\lfw\\most_images.xlsx")
print(df.head())

# new dataset dir
new_dataset_dir = "E:\\dataset\\lfw\\most_images\\"
```

```python
## Copy images to new dataset dir

if not os.path.exists(new_dataset_dir):
    os.makedirs(new_dataset_dir)
# copy people_folder to new dataset dir
import shutil

for index, row in df.iterrows():
    person_name = row["name"]
    src_folder = os.path.join(
        "E:\\dataset\\lfw\\lfw-deepfunneled\\lfw-deepfunneled", person_name
    )
    dst_folder = os.path.join(new_dataset_dir, person_name)
    if os.path.exists(src_folder):
        shutil.copytree(src_folder, dst_folder)
        print(f"Copied {person_name} to new dataset.")
    else:
        print(f"Source folder for {person_name} does not exist.")
```

```python
# Data Augmentation: Randomly flip images horizontally
```

```python
for person_name in os.listdir(new_dataset_dir):
    person_folder = os.path.join(new_dataset_dir, person_name)
    images = os.listdir(person_folder)
    num_images = len(images)
    if num_images < 50:
        images_to_augment = 50 - num_images
        for i in range(images_to_augment):
            img_name = random.choice(images)
            img_path = os.path.join(person_folder, img_name)
            img = Image.open(img_path)
            # Randomly choose augmentation method
            if random.random() < 0.5:
                # Horizontal flip
                augmented_img = img.transpose(Image.FLIP_LEFT_RIGHT)
            else:
                # Random rotation between -15 to 15 degrees
                angle = random.uniform(-15, 15)
                augmented_img = img.rotate(angle)
            # Save augmented image
            new_img_name = f"aug_{i}_{img_name}"
            new_img_path = os.path.join(person_folder, new_img_name)
            augmented_img.save(new_img_path)
        print(f"Augmented {images_to_augment} images for {person_name}.")
```
